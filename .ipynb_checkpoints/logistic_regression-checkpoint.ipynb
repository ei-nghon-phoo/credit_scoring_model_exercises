```python
# importing libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
```

```python
# Generate synthetic dataset
np.random.seed(42)
data_size = 1000

income = np.random.normal(50000, 15000, data_size)
age = np.random.normal(40, 10, data_size)
loan_amount = np.random.normal(10000, 5000, data_size)
credit_score = np.random.choice([0, 1], size=data_size, p=[0.3, 0.7])
```

```python
# Combine into a DataFrame
data = pd.DataFrame({
    'income': income,
    'age': age,
    'loan_amount': loan_amount,
    'credit_score': credit_score
})
```

```python
# Split the dataset into features (X) and target (y)
X = data[['income', 'age', 'loan_amount']]
y = data['credit_score']
```

```python
data.head(5)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize and train the logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

A confusion matrix is a table used to evaluate the performance of a classification algorithm. It provides a summary of the prediction results on a classification problem by showing the number of correct and incorrect predictions made by the model, compared to the actual outcomes (target values) in the dataset.
* TN (True Negative): The number of instances correctly predicted as negative.
* FP (False Positive): The number of instances incorrectly predicted as positive (Type I error).
* FN (False Negative): The number of instances incorrectly predicted as negative (Type II error).
* TP (True Positive): The number of instances correctly predicted as positive.


```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred, zero_division=0)
```

```python
# Plotting the confusion matrix
plt.matshow(conf_matrix, cmap=plt.cm.Blues, alpha=0.3)
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        plt.text(x=j, y=i, s=conf_matrix[i, j], ha='center', va='center')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

## example 2
with MM MFI context

```python
# Generate synthetic dataset
np.random.seed(42)
data_size = 1000

overlapped_MFI_count = np.random.randint(0, 5, data_size)
total_active_loan_count = np.random.randint(0, 10, data_size)
priciple_outstanding_amount = np.random.randint(10000 , 1000000, data_size)
loan_type = np.random.randint(0, 1, data_size) #0 means individual, 1 means group loan
business_type= np.random.randint(0, 8,data_size) #different type of businesses
credit_score = np.random.choice([0, 1], size=data_size, p=[0.3,0.7])
```

```python
# Combine into a DataFrame
data = pd.DataFrame({
    'overlapped_MFI_count': overlapped_MFI_count,
    'total_active_loan_count': total_active_loan_count,
    'priciple_outstanding_amount': priciple_outstanding_amount,
    'loan_type' : loan_type, 
    'business_type' : business_type,
    'credit_score': credit_score
})
```

```python
# Split the dataset into features (X) and target (y)
X = data[['overlapped_MFI_count', 'total_active_loan_count', 'priciple_outstanding_amount', 'loan_type','business_type']]
y = data['credit_score']
```

```python
data.head(5)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize and train the logistic regression model
model = LogisticRegression(max_iter = 1000)
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred, zero_division=0)
```

```python
# Plotting the confusion matrix
plt.matshow(conf_matrix, cmap=plt.cm.Blues, alpha=0.3)
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        plt.text(x=j, y=i, s=conf_matrix[i, j], ha='center', va='center')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
